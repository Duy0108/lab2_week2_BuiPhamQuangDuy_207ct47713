```python
import numpy as np
# Tạo mảng NumPy kích thước 150x5 với các giá trị ngẫu nhiên
# Chiều cao (cm) từ 150 đến 190
# Cân nặng (kg) từ 40 đến 100
# Tuổi từ 18 đến 24
# Lương (VND) từ 5,000,000 đến 30,000,000
# GPA từ 1.0 đến 4.0

data = np.column_stack((
    np.random.randint(150, 191, 150),  # Chiều cao
    np.random.randint(40, 101, 150),   # Cân nặng
    np.random.randint(18, 25, 150),    # Tuổi từ 18 đến 24
    np.random.randint(5000000, 30000001, 150),  # Lương (VND)
    np.random.uniform(1.0, 4.0, 150)   # GPA
))

# Hiển thị 5 dòng đầu tiên của mảng NumPy
print(data[:5])
```

```python
# Chia mảng thành X (bốn cột đầu tiên) và y (cột cuối cùng)
X = data[:, :-1]  # Lấy tất cả các dòng và bốn cột đầu tiên
y = data[:, -1]   # Lấy tất cả các dòng và chỉ cột cuối cùng

# Hiển thị X và y
print("X (Dữ liệu đầu vào):\n", X[:5])
print("y (Dữ liệu mục tiêu):\n", y[:5])
```

```python
from sklearn.model_selection import train_test_split
# Chia X thành X_train và X_test chứa 70% dữ liệu và chia y thành y_train và y_test, trong đó y_train chứa 70% dữ liệu
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Hiển thị kích thước của các tập dữ liệu
print("Kích thước của X_train:", X_train.shape)
print("Kích thước của X_test:", X_test.shape)
print("Kích thước của y_train:", y_train.shape)
print("Kích thước của y_test:", y_test.shape)
```

```python
# Tạo 10 tập dữ liệu không chồng chéo của X_train
X_train_splits = np.array_split(X_train, 10)

# Hiển thị kích thước của các tập dữ liệu con
for i, split in enumerate(X_train_splits):
    print(f"Kích thước của tập con X_train_{i+1}: {split.shape}")
```
