### **Bài tập 1: Tạo mảng và thao tác cơ bản**

1 Tạo một mảng NumPy với các giá trị từ 1 đến 20.\
2 Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng.\
3 Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100.\
4 Lấy hàng thứ 2 và cột thứ 3 của mảng 2D.

```python

import numpy as np

# 1. Tạo một mảng NumPy với các giá trị từ 1 đến 20
array_1 = np.arange(1, 21)
print("Mảng từ 1 đến 20:", array_1)

# 2. Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng
print("Tổng:", np.sum(array_1))
print("Giá trị lớn nhất:", np.max(array_1))
print("Giá trị nhỏ nhất:", np.min(array_1))
print("Trung bình:", np.mean(array_1))

# 3. Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100
array_2d = np.random.randint(0, 101, size=(3, 5))
print("Mảng 2D:", array_2d)

# 4. Lấy hàng thứ 2 và cột thứ 3 của mảng 2D
print("Hàng thứ 2:", array_2d[1, :])
print("Cột thứ 3:", array_2d[:, 2])

```

### **Bài tập 2: Các thao tác nâng cao**

1 Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1.\
2 Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1]).\
3 Tính tích vô hướng (dot product) của hai mảng 1D: [1, 2, 3] và [4, 5,6].\
4 Tạo một ma trận 5x5 và tính định thức (determinant) và nghịch đảo của ma trận.

```python

import numpy as np

# 1. Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1
random_array = np.random.rand(20)
print("Mảng ngẫu nhiên từ 0 đến 1:", random_array)

# 2. Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1])
# Vì mảng đã chứa giá trị trong khoảng [0, 1] nên việc chuẩn hóa không cần thiết.
# Tuy nhiên, nếu bạn muốn chuẩn hóa một mảng có giá trị ngoài khoảng [0, 1], có thể sử dụng:
# normalized_array = (random_array - np.min(random_array)) / (np.max(random_array) - np.min(random_array))
normalized_array = random_array  # Đây chỉ là mảng gốc vì đã trong khoảng [0, 1]
print("Mảng sau khi chuẩn hóa:", normalized_array)

# 3. Tính tích vô hướng (dot product) của hai mảng 1D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Tích vô hướng của a và b:", dot_product)

# 4. Tạo một ma trận 5x5 và tính định thức, nghịch đảo
matrix = np.random.rand(5, 5)  # Ma trận 5x5 với giá trị ngẫu nhiên
print("Ma trận:", matrix)

# Tính định thức của ma trận
determinant = np.linalg.det(matrix)
print("Định thức của ma trận:", determinant)

# Kiểm tra xem ma trận có nghịch đảo không (det != 0)
if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)  # Tính ma trận nghịch đảo
    print("Ma trận nghịch đảo:", inverse_matrix)
else:
    print("Ma trận không khả nghịch (định thức = 0).")
```

## **Phần 3: Làm sạch dữ liệu**

### **Bài tập 5: Xử lý dữ liệu thiếu**
1 Tạo một DataFrame chứa các giá trị sau:

| Name    | Age | City     | Salary |
|---------|-----|----------|--------|
| Alice   | 23  | New York | 60000  |
| Bob     | NaN | Boston   | 52000  |
| Charlie | 25  | NaN      | NaN    |
| David   | 24  | Chicago  | 58000  |
| Eva     | 22  | Boston   | NaN    |

2 Điền giá trị thiếu trong cột Age bằng giá trị trung bình.\
3 Xóa các hàng có nhiều hơn 1 giá trị thiếu.\
4 Điền giá trị thiếu trong cột Salary bằng 50000.

```python

import pandas as pd

# Tạo DataFrame chứa dữ liệu thiếu
data_with_missing = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, None, 25, 24, 22],
    'City': ['New York', 'Boston', None, 'Chicago', 'Boston'],
    'Salary': [60000, 52000, None, 58000, None]
}

df_missing = pd.DataFrame(data_with_missing)
print("Dữ liệu ban đầu:")
print(df_missing)

# 1. Điền giá trị thiếu trong cột Age bằng giá trị trung bình
mean_age = df_missing['Age'].mean()
df_missing['Age'].fillna(mean_age, inplace=True)
print("\nDữ liệu sau khi điền giá trị thiếu trong cột 'Age' bằng giá trị trung bình:")
print(df_missing)

# 2. Xóa các hàng có nhiều hơn 1 giá trị thiếu
df_missing.dropna(thresh=3, inplace=True)  # Thresh=3 có nghĩa là phải có ít nhất 3 giá trị không thiếu
print("\nDữ liệu sau khi xóa các hàng có nhiều hơn 1 giá trị thiếu:")
print(df_missing)

# 3. Điền giá trị thiếu trong cột Salary bằng 50000
df_missing['Salary'].fillna(50000, inplace=True)
print("\nDữ liệu sau khi điền giá trị thiếu trong cột 'Salary' bằng 50000:")
print(df_missing)

```

## **Phần 4: Trực quan hóa dữ liệu với Matplotlib**

### **Bài tập 6: Biểu đồ cơ bản**

1 Tạo một biểu đồ đường biểu diễn hàm số y = x ** 2 trên khoảng [-10,10] \
2 Vẽ biểu đồ cột thể hiện điểm số (Score) của các sinh viên từ Bài tập 3.\
3 Tạo một biểu đồ tròn (pie chart) thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris.

```python

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. Biểu đồ đường hàm số y = x^2 trên khoảng [-10,10]
x = np.linspace(-10, 10, 400)
y = x ** 2

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2", color='b')
plt.title("Biểu đồ hàm số y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# 2. Biểu đồ cột điểm số (Score) từ Bài tập 3
# Tạo DataFrame từ Bài tập 3
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, 25, 22, 24, 21],
    'Score': [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.bar(df['Name'], df['Score'], color='green')
plt.title("Biểu đồ cột điểm số của các sinh viên")
plt.xlabel("Tên sinh viên")
plt.ylabel("Điểm số")
plt.show()

# 3. Biểu đồ tròn (pie chart) thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris
# Tải dữ liệu Iris từ seaborn
iris = sns.load_dataset('iris')
flower_counts = iris['species'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(flower_counts, labels=flower_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title("Biểu đồ tròn tỷ lệ các loài hoa trong tập dữ liệu Iris")
plt.show()

```

### **Bài tập 7: Biểu đồ nâng cao**

1 Vẽ biểu đồ phân tán (scatter plot) giữa sepal_length và sepal_width của tập dữ liệu Iris. Dùng màu sắc để phân biệt các loại hoa (species).\
2 Thêm tiêu đề, nhãn trục và chú thích cho biểu đồ.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Tải dữ liệu Iris
iris = load_iris()
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_data['species'] = [iris.target_names[i] for i in iris.target]

# Biểu đồ phân tán
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=iris_data,
    x="sepal length (cm)",
    y="sepal width (cm)",
    hue="species",
    palette="viridis"
)

# Thêm tiêu đề, nhãn trục và chú thích
plt.title("Biểu đồ phân tán giữa Sepal Length và Sepal Width", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Sepal Width (cm)", fontsize=12)
plt.legend(title="Loại hoa", title_fontsize='13', fontsize='10', loc='upper right')

# Hiển thị biểu đồ
plt.show()
```
